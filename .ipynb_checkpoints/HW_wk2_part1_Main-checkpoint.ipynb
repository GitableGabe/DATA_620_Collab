In the context of the Amazon product co-purchasing network, having an edge from node 1 to node 15, it can be interpreted as: "Customers who bought product 1 often also bought product 15."


# Basic Analysis

Let's calculate the basic metrics of the graph


Create a graph from the edges

```python
# Create a directed graph
G = nx.from_pandas_edgelist(edges, source='FromNodeId', target='ToNodeId',
                            create_using=nx.DiGraph())

# Number of nodes, edges, average degree, and density
num_nodes = G.number_of_nodes()
num_edges = G.number_of_edges()
avg_degree = sum(dict(G.degree()).values()) / num_nodes
density = nx.density(G)

print(f"Number of nodes: {num_nodes}")
print(f"Number of edges: {num_edges}")
print(f"Average degree: {avg_degree:.2f}")
print(f"Density: {density:.6f}")
```



The graph is large with over 262,000 nodes and 1.23 million edges, indicating a complex network structure.

Let's find the graph's diameter


```python
# Find the graph diameter
try:
    diameter = nx.diameter(G)
except nx.NetworkXError:
    print("Graph is not strongly connected, diameter cannot be calculated directly.")
```

When we say that a graph is strongly connected and therefore its diameter cannot be calculated directly, it means that there are parts of the graph where not all nodes are reachable from every other node.

With that in mind it makes sense to subset the data and attempt to find the diameter of the subset.

the library random is utilized and a seed is set to ensure we retain the results of our first randomized subsets

```python
import random

# Set the random seed for reproducibility
random.seed(42)

# Number of nodes in the subset
subset_size = 4000

# Randomly select nodes
subset_nodes = random.sample(list(G.nodes()), subset_size)

# Create a subgraph from the subset of nodes
G_subset = G.subgraph(subset_nodes).copy()

# Plot the subgraph
plt.figure(figsize=(12, 8))
nx.draw(G_subset, with_labels=True, node_color="steelblue", node_size=400, font_weight='bold', edge_color="skyblue", alpha=0.5)
plt.show()
```

Things to note regarding above subset, because the nodes selected are random, the diameter would not be a good measure of the entire data set. This is because there is no telling how many more relationships would be established per node, if the entire data set is accounted for.

```python
# Find the graph diameter
try:
    diameter = nx.diameter(G)
except nx.NetworkXError:
    print("Graph is not strongly connected, diameter cannot be calculated directly.")
```

Regardless of the subset, the diameter does not seem calculable using solely **diameter()** method

This is because of the how strongly connected our directed graph is. Therefore, accounting for the connections among nodes, while ignoring the direction of the connections would be optimal. Networkx has two methods in its library that assists with this, **weakly_connected_components()** and **to_undirected()**.

I will attempt both to see if the diameter results differ.

```python
# Find the largest weakly connected component
largest_wcc_nodes = max(nx.weakly_connected_components(G_subset), key=len)
G_largest_wcc = G_subset.subgraph(largest_wcc_nodes).copy()

# Calculate the diameter of the largest weakly connected component
try:
    diameter_largest_wcc = nx.diameter(G_largest_wcc.to_undirected())
    print(f"The diameter of the largest weakly connected component is: {diameter_largest_wcc}")
except nx.NetworkXError:
    print("Graph is not strongly connected, diameter cannot be calculated directly.")
```

```python
# Convert the graph to undirected
G_undirected = G_subset.to_undirected()

if nx.is_connected(G_undirected):
    diameter = nx.diameter(G_undirected)
    print(f"The diameter of the graph is: {diameter}")
else:
    largest_cc = max(nx.connected_components(G_undirected), key=len)
    G_largest_cc = G_undirected.subgraph(largest_cc).copy()
    diameter = nx.diameter(G_largest_cc)
    print(f"The diameter of the largest connected component is: {diameter}")
```

As we can see, both methods for calculating the diameter without accounting for the direction, resulted in 2.

## Perform centrality measures on the Amazon product co-purchasing network

For the second analysis we will assess the centrality of the data. Centrality measures help identify the most important or influential nodes or products in the network. For instance, it can help determine products that are central to the network, which might be popular or highly influential in driving co-purchases.

We will compute the degree of centrality, the betweeness, and the closeness

```python
# Compute the degree of centrality
degree_centrality = nx.degree_centrality(G)

# Top 10 degree of centrality
degree_centrality_df = pd.DataFrame({
    'Degree Centrality': degree_centrality})

degree_centrality_df.sort_values(by='Degree Centrality',
                                 ascending=False).head(10)
```

These top 10 nodes with the highest degree of centrality mean these products are frequently co-purchased with many others. These products:

    Are highly influential in terms of driving co-purchase behavior.
    Can be targeted for promotional activities, as they have a wide-reaching influence.
    Could be central to recommendation algorithms, as their purchase can lead to a cascade of other purchases.
